# Analiza wyników sprzedaży

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('Global_Superstore2.csv', encoding='unicode_escape')
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')
```

## Sprzedaż według kategorii

```python
category_sales = df.groupby('Category', as_index=False)['Sales'].sum().sort_values('Sales', ascending=False)
plt.figure(figsize=(8, 4.5))
sns.barplot(data=category_sales, x='Category', y='Sales', palette='viridis')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales (USD)')
plt.tight_layout()
```

**Wnioski (PL):** Kategoria Technology generuje najwyższą sprzedaż, co sugeruje silny popyt na produkty technologiczne, podczas gdy Furniture i Office Supplies przynoszą wyraźnie niższe przychody.

**Findings (EN):** The Technology category delivers the highest sales, indicating strong demand for tech products, whereas Furniture and Office Supplies contribute substantially lower revenue.

## Zysk według regionów

```python
region_profit = df.groupby('Region', as_index=False)['Profit'].sum().sort_values('Profit', ascending=False)
plt.figure(figsize=(8, 4.5))
sns.barplot(data=region_profit, x='Region', y='Profit', palette='crest')
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
```

**Wnioski (PL):** Region APAC generuje najwyższy łączny zysk, podczas gdy Ameryka Południowa i Afryka notują straty, co wskazuje na potrzebę poprawy rentowności w tych obszarach.

**Findings (EN):** The APAC region yields the highest cumulative profit, whereas South America and Africa operate at a loss, signalling that profitability improvements are needed in those markets.

## Miesięczny trend sprzedaży

```python
monthly_sales = df.resample('M', on='Order Date')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 4.5))
sns.lineplot(data=monthly_sales, x='Order Date', y='Sales', marker='o', color='tab:blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Order Date')
plt.ylabel('Sales (USD)')
plt.tight_layout()
```

**Wnioski (PL):** Sprzedaż wykazuje sezonowe szczyty w czwartym kwartale każdego roku, co sugeruje nasilenie zakupów pod koniec roku i potrzebę odpowiedniego planowania zapasów.

**Findings (EN):** Sales display seasonal peaks in each year's fourth quarter, indicating intensified purchasing toward year-end and the need for appropriate inventory planning.
